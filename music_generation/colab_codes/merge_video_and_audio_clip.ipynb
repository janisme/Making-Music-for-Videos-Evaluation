```python
!pip --quiet install ftfy regex tqdm
!pip --quiet install git+https://github.com/openai/CLIP.git


import torch
import clip
from PIL import Image
device = "cuda" if torch.cuda.is_available() else "cpu"

print(f'Using device: {device}')
model, preprocess = clip.load("ViT-B/32", device=device)
def classify(im, labels):
    image = preprocess(im).unsqueeze(0).to(device)
    text = clip.tokenize(labels).to(device)

    with torch.no_grad():
        image_features = model.encode_image(image)
        text_features = model.encode_text(text)

        logits_per_image, logits_per_text = model(image, text)
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()
    max_prob = 0
    max_label = ""
    max_label_index = 0
    for i, label in enumerate(labels):
        # print(f"{labels[i]} = {probs[0,i]*100}%")
        if probs[0,i] > max_prob:
            max_prob = probs[0,i]
            max_label = labels[i]
            max_label_index = i
    return max_label, max_label_index
    # print(f"The most likely label is {max_label} with a probability of {max_prob*100}%")




```

```python

import os
import subprocess

mp4_files = [f for f in os.listdir() if f.endswith('.mp4')]

for mp4_file in mp4_files:
  new_filename = mp4_file.replace('.mp4', '_short.mp4')

  subprocess.run(['ffmpeg', '-i', mp4_file, '-t', '30', '-c:v', 'libx264', '-c:a', 'aac', new_filename])
```

```python

import glob

mp4_files = glob.glob("*_short.mp4")

print(mp4_files)
```

```python

import re

def get_number_from_filename(filename):
  match = re.search(r'\d+', filename)
  if match:
    return int(match.group(0))
  else:
    return 0

sorted_mp4_files = sorted(mp4_files, key=get_number_from_filename)

mp4_files = sorted_mp4_files
print(len(mp4_files))
print(mp4_files)
```

```python

import glob

mp3_files = glob.glob("*.mp3")

print(mp3_files)
```

```python

labels = [file_name[:-4] for file_name in mp3_files]
labels

```

```python
!pip install decord
import decord

from decord import VideoReader

music_list = []
label_list = []
for video_fname in mp4_files:
  vr = VideoReader(video_fname)
  frame = vr[100]
  frame_values = frame.asnumpy()
  im = Image.fromarray(frame_values)
  max_label, max_label_idx = classify(im, labels)
  music_list.append(mp3_files[max_label_idx])
  label_list.append(max_label)
```
