```python

import re

tuples = []

for file in mp3_files:
  numbers = re.findall(r'\d+', file)

  tuple = (int(numbers[0]), int(numbers[1]), int(numbers[2]))

  tuples.append(tuple)

print(tuples)
```

```python

tuples_with_indices = []

for i, tuple in enumerate(tuples):
  tuples_with_indices.append((tuple, i))

sorted_tuples_with_indices = sorted(tuples_with_indices, key=lambda x: x[0][0])

for tuple, index in sorted_tuples_with_indices:
  print(f"Tuple: {tuple}, Original index: {index}")
```

```python

from moviepy.editor import VideoFileClip, AudioFileClip, concatenate_audioclips

# change the file name from orange.mp4 to origin.mp4
video_path = "origin.mp4"
video_clip = VideoFileClip(video_path)

audio_clips = []


audio_files = []
iddx = 0
real_start_times = [0,
3.871,
10.844,
16.817,
19.086,
20.888,
26.793,
30.831,
43.443,
45.045,
49.383]
for tuple, index in sorted_tuples_with_indices:
    print(f"Tuple: {tuple}, Original index: {index}")
    audio_files.append((mp3_files[index], real_start_times[iddx]))
    iddx += 1
# for i in range(len(audio_files)):



for audio_file, start_time in audio_files:
    audio_clip = AudioFileClip(audio_file)
    audio_clip = audio_clip.set_start(start_time)
    audio_clips.append(audio_clip)


final_audio = concatenate_audioclips(audio_clips)

video_clip = video_clip.set_audio(final_audio)

output_path = "precise_output_video.mp4"
video_clip.write_videofile(output_path, codec="libx264", audio_codec="aac")


video_clip.close()
final_audio.close()


```
